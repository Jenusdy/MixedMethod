```python
with open('/data/ksa/04_Data_Preprocessing/32/01_imputation/48MYT_imputed_data.pkl','rb') as file:
    df=pickle.load(file)
```

```python
df_idpoint=df.query('idpoint=="320431003C2#08"')
```

```python
df_idpoint['Sigma0_VH_db']=df_idpoint.apply(lambda y: np.nan if y.weight==0 else y.Sigma0_VH_db_imputation, axis=1)
df_idpoint['Sigma0_VV_db']=df_idpoint.apply(lambda y: np.nan if y.weight==0 else y.Sigma0_VV_db_imputation, axis=1)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Example data
# Create a plot
plt.figure(figsize=(10, 6))
data = df_idpoint.Sigma0_VV_db_imputation
plt.plot(data, marker='o', linestyle='-', color='r', label='VV imputation')

data = df_idpoint.Sigma0_VV_db
plt.plot(data, marker='o', linestyle='-', color='b', label='VV original')

# Add title and labels
plt.title('Simulation of Imputation')
plt.xlabel('Index')
plt.ylabel('Values')
# Add grid and legend
plt.grid(True)
plt.legend()
# Show the plot
plt.show()
```
