```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# Sample data for the pie charts
data = [
    [35, 25, 25, 15],
    [50, 10, 20, 20],
    [20, 30, 30, 20],
    [15, 30, 35, 20],
]

labels = ['A', 'B', 'C', 'D']

# Create a figure
fig = plt.figure(figsize=(10, 6))

# Create a 2x2 grid of subplots
gs = GridSpec(2, 2, figure=fig)

# Create pie charts in the grid
for i in range(4):
    ax = fig.add_subplot(gs[i])
    ax.pie(data[i], labels=labels, autopct='%1.1f%%')
    ax.set_title(f'Pie Chart {i + 1}')

# Adjust layout

plt.legend()
plt.tight_layout()
# Show the plot
plt.show()
```

```python
import numpy as np

def standardize_3d_array(arr,axs):
    # Calculate mean and std along the second axis (axis=1)
    mean = np.mean(arr, axis=axs, keepdims=True)
    std = np.std(arr, axis=axs, keepdims=True)
    
    # Standardize along the second axis
    print('mean')
    print(mean)
    print('std')
    print(std)
    standardized_arr = (arr - mean) / std
    return standardized_arr
```
