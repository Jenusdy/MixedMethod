```python
import pandas as pd

# Replace with the path to your CSV file
file_path = '/data/ksa/00_Data_Input/relabelled_data_ksa_explore.csv'

# Reading the CSV file
data = pd.read_csv(file_path)

# Displaying the first few rows of the data
data.head()
```

```python
import matplotlib.pyplot as plt

# Menghitung frekuensi setiap kategori untuk dua variabel
frequencies_class = data['class'].value_counts()
frequencies_bulan = data['bulan'].value_counts()

# Membuat subplots
fig, axes = plt.subplots(1, 2, figsize=(8, 3))

# Membuat barchart untuk 'class'
axes[0].bar(frequencies_class.index, frequencies_class.values)
axes[0].set_title('Class Distribution')
axes[0].set_xlabel('Class')
axes[0].set_ylabel('Frequency')

# Membuat barchart untuk 'bulan'
axes[1].bar(frequencies_bulan.index, frequencies_bulan.values)
axes[1].set_title('Month Distribution')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Frequency')

# Menampilkan barchart
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Split 'periode' into 'year' and 'period'
data[['year', 'period']] = data['periode'].str.split('_', expand=True)

# Convert 'year' and 'period' to integers for proper sorting
data['year'] = data['year'].astype(int)
data['period'] = data['period'].astype(int)

# Sort the data by 'year' and 'period'
data = data.sort_values(by=['year', 'period'])

# Iterate over each VH column
for i in range(31):  # Assuming VH_0 to VH_30
    vh_column = f'VH_{i}'
    
    # Create the FacetGrid for separate boxplots
    g = sns.FacetGrid(data, col="class", col_wrap=4, height=4, aspect=1.5)
    g.map(sns.boxplot, "periode", vh_column, order=sorted(data['periode'].unique()))
    
    # Rotate x-axis labels and enable grid on each subplot
    for ax in g.axes.flat:
        ax.grid(True)  # Enable grid for each subplot
        for label in ax.get_xticklabels():
            label.set_rotation(90)
    
    g.add_legend()
    g.set_axis_labels("Periode", vh_column)
    plt.subplots_adjust(top=0.9)
    g.fig.suptitle(f'Boxplot for {vh_column}')  # Add a title to the grid
    plt.show()

# Optional: Drop the 'year' and 'period' columns if they are no longer needed
data = data.drop(columns=['year', 'period'])
```

```python
# Define the column names for VH_0 to VH_30
vh_columns = [f'VH_{i}' for i in range(31)]

# Define the column name for the class/label
class_column = 'class'  # Replace 'class' with the actual name of your class/label column

# Get unique classes
unique_classes = data[class_column].unique()

# Set up the subplot grid
fig, axes = plt.subplots(2, 4, figsize=(10, 10))  # Adjust size as needed
axes = axes.flatten()  # Flatten the 2D array to 1D for easier indexing

# Calculate and plot correlation matrix for each class
for i, class_value in enumerate(unique_classes):
    # Filter data for the current class
    class_data = data[data[class_column] == class_value]
    
    # Subset data for VH_0 to VH_30
    vh_data = class_data[vh_columns]
    
    # Calculate the correlation matrix
    correlation_matrix = vh_data.corr()
    
    # Plot the correlation matrix
    sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', fmt=".2f", linewidths=0.5, ax=axes[i])
    axes[i].set_title(f'Class: {class_value}')
    
# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```

```python
from matplotlib.ticker import FuncFormatter

# Split 'periode' into 'year' and 'period'
data[['year', 'period']] = data['periode'].str.split('_', expand=True)

# Convert 'year' and 'period' to integers for proper sorting
data['year'] = data['year'].astype(int)
data['period'] = data['period'].astype(int)

# Sort the data by 'year' and 'period'
data = data.sort_values(by=['year', 'period'])

# Function to format y-axis labels to 2 decimal places
def format_func(value, tick_number):
    return f'{value:.2f}'

# List of F columns
f_columns = [f'F{i}' for i in range(1, 18)]  # F1 to F17

# Iterate over each F column
for f_column in f_columns:
    
    # Create the FacetGrid for separate boxplots
    g = sns.FacetGrid(data, col="class", col_wrap=4, height=4, aspect=1.5)
    g.map(sns.boxplot, "periode", f_column, order=sorted(data['periode'].unique()))
    
    # Rotate x-axis labels and enable grid on each subplot
    for ax in g.axes.flat:
        ax.grid(True)  # Enable grid for each subplot
        ax.yaxis.set_major_formatter(FuncFormatter(format_func))  # Set y-axis format
        for label in ax.get_xticklabels():
            label.set_rotation(90)
    
    g.add_legend()
    g.set_axis_labels("Periode", f_column)
    plt.subplots_adjust(top=0.9)
    g.fig.suptitle(f'Boxplot for {f_column}')  # Add a title to the grid
    plt.show()

# Optional: Drop the 'year' and 'period' columns if they are no longer needed
data = data.drop(columns=['year', 'period'])
```
